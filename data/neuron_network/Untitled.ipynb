```python
import pandas as pd
import string
```

```python
df_test = pd.read_csv("SK하이닉스.csv")
```

```python
df_test_p1 = df_test['pattern1']
df_test_p1
```

```python
df = pd.DataFrame({"1P":[],
                   "1M":[],
                   "1K":[],
                   "20":[],
                   "21":[],
                   "30":[],"31":[],"32":[],"33":[],"34":[],"35":[]})
```

```python
df["1P"] = [1.0 if temp[0] == "P" else 0.0 for temp in df_test['pattern1']]
```

```python
df = df.fillna(0)
```

```python
for i in range(len(df)):
    temp = df_test['pattern1'][i]
    tmp1 = temp[0]
    tmp2 = temp[1]
    tmp3 = temp[2]
    if tmp1 == "M":
        df['1M'].iloc[i] = 1
    elif tmp1 == "K":
        df['1K'].iloc[i] = 1
        
    if tmp2 == "0":
        df['20'].iloc[i] = 1
    elif tmp2 == "1":
        df['21'].iloc[i] = 1
        
    if tmp3 == "0":
        df['30'].iloc[i] = 1
    elif tmp3 == "1":
        df['31'].iloc[i] = 1
    elif tmp3 == "2":
        df['32'].iloc[i] = 1
    elif tmp3 == "3":
        df['33'].iloc[i] = 1
    elif tmp3 == "4":
        df['34'].iloc[i] = 1
    elif tmp3 == "5":
        df['35'].iloc[i] = 1
```
